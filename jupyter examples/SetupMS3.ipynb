# Setup input files and mesh for sycamore simulations

```python
import sys
import os
import matplotlib.pyplot as plt
from datetime import datetime

extra_path = "/Users/lraming/Documents/Repos/GP4"
if extra_path not in sys.path:
    sys.path.append(extra_path)
    
from tribsmodel import Model as model
m = model()

os.chdir("../..")
```

__read in template.in and setup relevant vars__

```python
m.read_input_file("src/infiles/template.in")
basin = "sycMS3"
base_name = "ms3_"

# todays_date = datetime.now()
# todays_date = todays_date.strftime("%Y-%m-%d")

todays_date = "2023-10-24"

results_base_path = 'results/'+basin+'/'+todays_date+'/'
results_base_hyd_path = results_base_path+'/hyd/'
```

__create relevant directories if needed__

```python
## USE this to add subdirectories if they do not exist
# Initialize the base directory
directories = results_base_hyd_path.split('/')
results_base_hyd_path = ''

# Create the directories one by one
for directory in directories:
    results_base_hyd_path = os.path.join(results_base_hyd_path, directory)
    try:
        os.mkdir(results_base_hyd_path)
    except FileExistsError:
        # Directory already exists, so continue without raising an error
        pass
```

__modify input options for individual simulations__

```python
points = "MS3s1z1_tree6_1.points"
stations = "Rain_MS_3.sdf"
graphfile = "sycMS3.in_flow_4nodes.reach"

# opt1:"sycMS3.in_flow_16nodes.reach"
# opt2: "sycMS3.in_nconn_4nodes.reach"
# opt3:"sycMS3.in_upnconn_4nodes.reach"
```

```python
m.options['pointfilename']['value'] = "data/model/points/"+points
m.options['gaugestations']['value'] = "data/model/rain/"+stations
m.options['graphfile']['value'] = 'data/model/parallel_mesh/'+basin+'/'+graphfile
m.options['graphoption']['value'] = 1
m.options['outfilename']['value'] = results_base_path+base_name
m.options['outhydrofilename']['value'] = results_base_hyd_path+base_name
m.options['gwaterfile']['value'] = "data/model/Sycamore.iwt"


# this might already be set
m.options["soiltablename"]["value"] = 'data/model/2023-10-16/Sycamore_new.sdt'
```

```python
m.check_paths()
```

```python
m.write_input_file("src/infiles/sycMS3.in") # save to infiles
```

```python
m.options["runtime"]["value"] = 0
m.write_input_file("sycMS3.in") # try local run just to build mesh
```

```python
m.run('./tRIBS','sycMS3.in',mpi_command='mpirun -n 5')
```

```python
m.geo["EPSG"] = "EPSG:26912" 
# read in voi for #2 processor
data_dir = "results/sycMS3/2023-10-24"
voi_files = [f for f in os.listdir(data_dir) if 'voi.' in f]
# Create a single figure and axis
fig, ax = plt.subplots(figsize=(10, 10))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  


for i, file in enumerate(voi_files):
    # try:
    print(file)
    shapes = m.read_voi_file(data_dir+'/'+file)
    if shapes is not None:
        voi = shapes[0]
        voi.plot(ax=ax,color=colors[i % len(colors)],label = file)
    # except ValueError:
    #     print(ValueError())

m.options['outfilename']['value'] = "results/sycMS3/2023-10-24/"+base_name
reaches = m.read_reach_file()
reaches.plot(ax = ax, color = 'k')
```
